```python

open_apc_urls = ["https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv",
                 "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2016.csv",
                 "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2016_enriched.csv",
                 "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2018.csv",
                 "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2018_enriched.csv"]

i = 0
apcs = [0,1,2,3,4]
for url in open_apc_urls:
    idx = i
    apc_file = "openapcs"+ str(idx) +".csv"
    urllib.request.urlretrieve(url, apc_file)
    apcs[idx] = pd.read_csv(apc_file)
    colNames = apcs[idx].columns
    colNames.contains("brutto (euro)")
    if colNames.contains("brutto (euro)"):
        tmp = apcs[idx]["brutto (euro)"].replace(',', '')
        euroSeries = pd.to_numeric(tmp, errors='coerce')
        #euroSeries.hist
    else:
        apcs[idx]["euro"]
        apcs[idx]["euro"].hist(alpha = 0.2)
    i = i+1

plt.show()
euroSeries
```
